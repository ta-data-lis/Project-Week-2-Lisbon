```python
ds_elections=elections
ds_elections=ds_elections.drop(columns=['a_wahlkreis','b_wahlkreis','bezirkId','ost_west'])
ds_elections.head()
```

```python
collapse_election=ds_elections.groupby(['bezirk'], as_index=False).agg({'wahlberechtigte':'sum','waehler':'sum','gueltig':'sum','ungueltig':'sum','CDU':'sum','DIE LINKE':'sum','SPD':'sum','GRUNE':'sum','FDP':'sum','PIRATEN':'sum','NPD':'sum','BuSo':'sum','MLPD':'sum','AfD':'sum','BIG':'sum','pro Deutschland':'sum','FREIE WAHLER':'sum','Die PARTEI':'sum','Boes':'sum','B':'sum','Bundnis 21_RRP':'sum','DKP':'sum','Die Violetten':'sum','Ditsche':'sum','Ditsche':'sum','Di Leo':'sum','Sylla':'sum','Fricke':'sum','Otto':'sum','Beckmann':'sum','Snelinski':'sum'})
collapse_election.head(15)
```

```python
small_parties=collapse_election.drop(columns=['bezirk','wahlberechtigte','waehler','gueltig','ungueltig','CDU','DIE LINKE','SPD','GRUNE','AfD'])
small_parties_sum=small_parties.sum(axis=1)
small_parties_sum
```

```python
parties_and_small=collapse_election[['bezirk','wahlberechtigte','waehler','gueltig','ungueltig','CDU','DIE LINKE','SPD','GRUNE','AfD']]
parties_and_small['Others']=small_parties_sum
parties_and_small.head(15)
```

```python
percentage_table=parties_and_small
#summe_wahlberecht=percentage_table.sum(['wahlberechtigte'])
#percentage_table['wahlberechtigte']=percentage_table['wahlberechtigte']/summe_wahlberecht
#percentage_table
#summe_wahlberecht
percentage_table['CDU']=percentage_table['CDU']*100/percentage_table['gueltig']
percentage_table['DIE LINKE']=percentage_table['DIE LINKE']*100/percentage_table['gueltig']
percentage_table['SPD']=percentage_table['SPD']*100/percentage_table['gueltig']
percentage_table['GRUNE']=percentage_table['GRUNE']*100/percentage_table['gueltig']
percentage_table['AfD']=percentage_table['AfD']*100/percentage_table['gueltig']
percentage_table['Others']=percentage_table['Others']*100/percentage_table['gueltig']
percentage_table
```

```python
percentage_table['ungueltig']=percentage_table['ungueltig']*100/percentage_table['waehler']
percentage_table['gueltig']=percentage_table['gueltig']*100/percentage_table['waehler']
```

```python
percentage_table['waehler']=percentage_table['waehler']*100/percentage_table['wahlberechtigte']
percentage_table.head()
```
