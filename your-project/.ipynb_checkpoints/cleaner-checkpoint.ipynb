```python
import pandas as pd
import numpy as np
```

```python
#NAME OF DATA-SET TO CLEAN
#Take into account FILE PATH as WELL
path = "DataSet/2017_densitat.csv"
```

```python
#Import file
data = pd.read_csv(path)
data
```

```python
#DROP DUPLICATES
before_data = len(data)
data_nodup = data.drop_duplicates()
after_data = len(data_nodup)
print('Number of duplicate records dropped: ', str(before_data - after_data))

data = data.drop_duplicates()
```

```python
#CHANGE RELEVANT COLUMN NAMES (translate if not in English) IF NEEDED
col_names = ["Year","Distr-Code","District-Name","Neigh-Name",
             "Neigh-Name","Population","Area (ha)","Residential Area (ha)","Density (hab/ha)",
             "Net Density (hab/ha)"]
data.columns = col_names
```

```python
#EXTRACT RELEVANT PARAMETERS by RELEVANT COLUMN
result = data.groupby('District-Name', as_index=True).agg({'Population':'sum','Area (ha)':'sum',
                                                          "Residential Area (ha)":"sum",
                                                          "Density (hab/ha)":"sum",
                                                           "Net Density (hab/ha)":"sum"})
result
```

```python
#ADD REFERENCE YEAR OF DATA
year = [2017]*len(result["Population"])
result["Year"] = year
result
```

```python
#TRANSPOSE DATA
result = result.transpose()
result
```

```python
#EXPORT TO csv FILE WITH APPROPRIATE NAME and PATH

result.to_csv('DataSet/clean-pop-density.csv', index=True)
```
